### VIF ( Remove Multicollinearity)

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# removed high vif value
new_df.drop(columns=['carlength','carwidth','carheight'],inplace=True)
new_df.drop(columns=['wheelbase'],inplace=True)
new_df.drop(columns=['range_Medium cost cars'],inplace=True)
new_df.drop(columns=['fuelsystem_mpfi'],inplace=True)
new_df.drop(columns=['drivewheel_fwd'],inplace=True)
new_df.drop(columns=['boreratio'],inplace=True)
new_df.drop(columns=['carbody_wagon'],inplace=True)
```

```python
def cal_vif(dataset):
    vif=pd.DataFrame()
    vif['features']=dataset.columns
    vif['vif_value']= [variance_inflation_factor(dataset.values,i) for i in range(dataset.shape[1])]
    
    return vif
```

```python
features = new_df.drop(columns=['price'])
vif_result=cal_vif(features)
```
